```
import re

file_list = [ 'mallika_1.jpg', 'dog005.jpg', 'grandson_2018_01_01.png', 'dog008.jpg', 'mallika_6.jpg', 'grandson_2018_5_23.png', 'dog01.png', 
             'mallika_11.jpg', 'mallika2.jpg', 'grandson_2018_02_5.png', 'grandson_2019_08_23.jpg', 'dog9.jpg', 'mallika05.jpg' ]


def files_sort(files):
  files_type = [fname.split(".") for fname in files]
  def name_descr_split(fname):
    match = re.compile("[a-zA-Z]+").search(fname)
    return ((fname[:match.end()], fname[match.end():]))

  files_split = [(*name_descr_split(fname[0]),fname[1]) for fname in files_type]
  files_split = [(fname[0],fname[1].strip("_").split("_"),fname[2]) for fname in files_split]
  files_split.sort(key = lambda fname : [fname[0],int("".join(["{:0>2}".format(i) for i in fname[1]]))])
  files_split = [(fname[0],'_'.join(fname[1]),fname[2]) for fname in files_split]
  files_split = [("_".join(fname[0:2]),fname[2]) for fname in files_split]
  files_split = [".".join(fname) for fname in files_split]
  return files_split

files_sort(file_list)
```
